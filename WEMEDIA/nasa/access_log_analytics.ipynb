- Find the top-10 (host) hosts makes most requests from 18th Aug to 20th Aug
- Find out the country with most requests originating from (according the source host / IP)

```python
import pandas as pd
```

```python
file = './access_log_Aug95_short'
access_log = pd.read_csv(file, sep=' ', header=None)
access_log.head()
```

```python
drop_cols = [1,2,4]
access_log.drop(drop_cols, axis=1)
```

```python
new_cols = {0:'host', 3:'timestamp', 5:'request', 6:'HTTP', 7:'bytes' }
access_log.drop(drop_cols, axis=1).rename(columns=new_cols)
```

```python
# Count the total number of HTTP requests recorded by this access logfile
len(access_log.drop(drop_cols, axis=1).rename(columns=new_cols))
```

```python
# Find the top-10 (host) hosts makes most requests from 18th Aug to 20th Aug
access_log.drop(drop_cols, axis=1).rename(columns=new_cols).groupby('host').size().nlargest(10)
```

```python
# Find out the country with most requests originating from (according the source host / IP)
import requests
ip = '133.43.96.45'
r = requests.get('http://ipinfo.io/%s/json' % ip)
r.json()['country']
```
